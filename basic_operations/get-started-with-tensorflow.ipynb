```python
# Necessary imports

import tensorflow as tf
```

### 1. Create a vector, scalar, matrix and tensor with values using `tf.constant()` ###

```python
scalar = tf.constant(5)
vector = tf.constant([1, 2])
matrix = tf.constant([[1, 2], [3, 4]])
tensor = tf.constant([[[1,2], [3,4]], [[5, 6], [7,8]]])
```

### 2. Find the shape of the tensors created in 1. ###

```python
print(f"Scalar's shape: {scalar.shape}")
print(f"Vector's shape: {vector.shape}")
print(f"Matrix's shape: {matrix.shape}")
print(f"Tensor's shape: {tensor.shape}")
```

### 3. Create two tensors containing random values between 0 and 1 with shape [5, 300]. ###

```python
t1 = tf.random.uniform(shape = (5,300), minval = 0, maxval = 1.0)
t2 = tf.random.uniform(shape = (5,300), minval = 0, maxval = 1.0)

t2[:5]
```

### 4. Multiply the two tensors created in 3 using matrix multiplication. ###

```python
t3 = tf.matmul(tf.transpose(t1), t2)
t3[:5]
```

### 5. Multiply the two tensors created in 3 using dot product. ###

```python
t4 = tf.tensordot(tf.transpose(t1), t2, axes = 1)
t4[:5]
```

### 6. Create a tensor with random values between 0 and 1 with shape [224, 224, 3]. ###

```python
t5 = tf.random.uniform(shape =  [224, 224, 3], minval = 0, maxval = 1)
```

### 7. Find the min and max values of the tensor in 6. ###

```python
print(f"Min value: {tf.math.reduce_min(t5)}")
print(f"Max value: {tf.math.reduce_max(t5)}")
```

### 8. Create a tensor with random values of shape [1, 224, 224, 3] then squeeze it to change the shape to [224, 224, 3]. ###

```python
t6 = tf.random.uniform(shape = [1, 224, 224, 3], minval = 0, maxval = 1)

t6 = tf.squeeze(t6)
print(t6.shape)
```

### 9. Create a tensor with shape [10], then find the index which has the maximum value. ###

```python
t7 = tf.range(0, 11)

print(tf.argmax(t7))
```

### 10. One-hot encode the tensor created in 9. ###

```python
tf.one_hot(t7, depth = 10)
```
